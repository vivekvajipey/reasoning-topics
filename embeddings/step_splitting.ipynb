```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('data/cot_data_50_start_sentences_sentsplit.csv')
df['CoT Sentences'] = df['CoT Sentences'].apply(ast.literal_eval)
```

```python
for response in df['CoT Response'].tolist():
    print(response)
```

```python
import re

def split_sentences(text):
    # Define a pattern that splits on a period (not followed by a digit) followed by a space or a newline
    sentences = re.split(r'(?<!\d)\.(?:\s|\n)|\n', text)
    # Remove any empty strings that may have been added in the list
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
    return sentences

# Apply the function to the 'CoT Response' column
df['CoT Sentences'] = df['CoT Response'].apply(split_sentences)
```

```python
df['CoT Length'] = df['CoT Sentences'].apply(len)
```

```python
df.to_csv('data/cot_data_50_start_sentences_sentsplit.csv', index=False)
```
