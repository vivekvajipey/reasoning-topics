```python
import os

log_directory = "logs"
valid_paths = []

for filename in os.listdir(log_directory):
    filepath = os.path.join(log_directory, filename)
    if os.path.getsize(filepath) > 0:
        valid_paths.append(filename)
```

```python
import pandas as pd
import re
import os

log_files = ['logs/' + path for path in valid_paths]

df = pd.DataFrame(columns=['iters', 'min_cf', 'min_df', 'rm_top', 'num_topics', 'num_beta_sample', 'removed_top_words', 'topics'])

def extract_topics_from_log(log_file_path):
    topics = []
    start_collecting = False

    with open(log_file_path, 'r') as file:
        for line in file:
            if line.startswith("Topic #"):
                start_collecting = True
            if start_collecting:
                if line.startswith("Topic #"):
                    topics.append(line.strip())
                else:
                    break

    return '\n'.join(topics)

def parse_log_file(log_file):
    filename = os.path.basename(log_file)
    params = re.findall(r"-([a-z_]+)-(\d+)", filename)

    num_beta_sample_match = re.search(r"-(\d+)\.log$", filename)
    if num_beta_sample_match:
        params.append(('num_beta_sample', num_beta_sample_match.group(1)))

    param_dict = {param: int(value) for param, value in params}

    with open(log_file, 'r') as file:
        content = file.read()
        removed_top_words = re.search(r"Removed Top words: (.+?)\n", content)

        param_dict['removed_top_words'] = removed_top_words.group(1) if removed_top_words else ""
    
    param_dict['topics'] = extract_topics_from_log(log_file)

    return param_dict

# List to hold new rows
new_rows = []

# Parse each log file and append to new_rows
for log_file in log_files:
    row = parse_log_file(log_file)
    new_rows.append(row)

# Create a DataFrame from new_rows and concatenate it with the original DataFrame
df_extended = pd.DataFrame(new_rows)
df = pd.concat([df, df_extended], ignore_index=True)
```
