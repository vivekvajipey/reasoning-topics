```python
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cot_df = pd.read_csv('data/cot_data_50_start_sentences.csv')
```

```python
cot_df.head()
```

```python
cot_df['CoT Sentences'] = cot_df['CoT Sentences'].apply(ast.literal_eval)
sentence_lengths = cot_df['CoT Sentences'].apply(len)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(sentence_lengths, bins=range(1, sentence_lengths.max()+1), align='left', color='skyblue', edgecolor='black')
plt.title('Frequency of Sentence Lengths in CoT Responses (GSM8K First 50, GPT-3.5)')
plt.xlabel('Number of Sentences in CoT Response')
plt.ylabel('Frequency')
plt.xticks(range(1, sentence_lengths.max()+1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```
